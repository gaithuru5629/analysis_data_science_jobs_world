#### Data Cleaning
The problems or things that we need to address in the data cleaning are:

1. The data set contains the following empty cells by column:
        a. JobTitle  has (65670-56603) empty cells
        b. Company has (65670-23459) empty cells
        c. CompanyReviews has (65670-27705) empty cells. 
   Remove rows containing empty cells. Also remove rows with a country field only.
2. The data type for all the columns is 'object'. We need to convert the data types of the columns as follows:
        a. JobTitle should be 'string'
        b. Company should be 'string'
        c. Location should be 'string'
        d. Link should be 'string'.
        e. Reviews should be 'int'. Prior to that, remove the word "reviews" in each cell.
3. For the Location column, split it into three columns:
        a. Country
        b. City 
        c. Postal Code
   Ensuring that rows with either one of them are accounted for.
     * Add country to corresponding rows.
     
     

```python
#deleting empty rows to avoid errors
dsJobs_df.dropna(inplace=True)
```

```python
dsJobs_df.info()
```

```python
dsJobs_df.head(20)
```

```python
#specifying data types of columns
new_dtypes = {'JobTitle':str, 'Company':str, 'Location':str, 'Reviews':str, 'Link':str}
dsJobs_df_1 = dsJobs_df.astype(new_dtypes)
dsJobs_df_1.dtypes
```

```python
dsJobs_df_1.info()
```

### Cleaning Reviews Column
By removing \n in all the cells

```python
#Removing \n in Company
dsJobs_df['Company'] = dsJobs_df['Company'].map(lambda x: str(x)[5:])
dsJobs_df.head()
```

```python
dsJobs_df['Reviews'].str.rsplit(' ', 1).str.get(0)

#rstrip() to remove trailing space
#lstrip() to remove leading space
#strip() to remove leading and trailing space
```

```python
#Remove the word "reviews" in review column
dsJobs_df['Reviews'] = dsJobs_df['Reviews'].map(lambda x: str(x)[0:-8])
dsJobs_df.head()
```

```python
#Convert the Reviews column to int data type.
#First remove comma
dsJobs_df.Reviews = dsJobs_df.Reviews.apply(lambda x: x.replace(',',''))
#dsJobs_df.head(10)
```

```python
dsJobs_df.head(10)
```

```python
#To remove spaces in between numbers in Reviews column
dsJobs_df['Reviews'] = dsJobs_df['Reviews'].str.replace(" ","")  #remove leading, trailing and all spaces in between
```

```python
dsJobs_df.tail()
```

```python
dsJobs_df.head()
```

### Converting Reviews Data Type to Integer

```python
#Then convert to int
dsJobs_df['Reviews'] = pd.to_numeric(dsJobs_df['Reviews'], errors='coerce')
dsJobs_df['Reviews'].dtypes
```

```python
dsJobs_df.head()
```

```python
dsJobs_df.info()
```

```python
#Removing null value cells created when the coercion to float data type failed for certain review cells
dsJobs_df1 = dsJobs_df.dropna(inplace=False)
dsJobs_df1
```

```python
#Converting company to string
dsJobs_df1['Company'] = dsJobs_df['Company'].astype('str')
dsJobs_df1['Company'].dtypes
```

Note: the object datatype is either string, or a mixture of numeric and non-numeric values

### Cleaning the JobTitle column
By deleting all text in brackets

```python
dsJobs_df1['JobTitle'] = dsJobs_df1['JobTitle'].str.split('(').str[0]
dsJobs_df1
```

### Cleaning the Location column so that we only have the city name
By deleting all text after the comma, if any

```python
dsJobs_df1["City"] = dsJobs_df1["Location"].str.split(',').str[0]
```

```python
dsJobs_df1.info()
```

```python
dsJobs_df1.head()
```

#### Finding and Remove any Duplicates

```python
#identifying if there are any duplicates
dsJobs_df1.duplicated()
```

```python
#removing duplicates if any
dsJobs_df2 = dsJobs_df1.drop_duplicates(inplace=False)
dsJobs_df2.info()
```

```python
#Taking a subset of the DataFrame, excluding the location column
dsJobs = dsJobs_df2[['JobTitle', 'Company', 'City', 'Reviews']]
dsJobs.head()
```

```python
#Resetting the index
dsJobs.reset_index(inplace=True)

dsJobs = dsJobs[['JobTitle', 'Company', 'City', 'Reviews']]
dsJobs
```

#### How many distinct job positions are there and how many of each position

```python
dsJobs_grouped1 = dsJobs_sorted.groupby(['JobTitle']).Company.count()
dsJobs_grouped1 = dsJobs_grouped1.reset_index()
print("Number of job titles related to data science, across all companies")
dsJobs_grouped1_1 = dsJobs_grouped1.drop(0)
top_ten_jobs = dsJobs_grouped1_1.head(11)
top_ten_jobs
```

```python
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("#733","")
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("#738","")
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("#740","")
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("#741","")
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("#14545 - ","")
top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.replace("17-18 ","")
#top_ten_jobs['JobTitle'] = top_ten_jobs['JobTitle'].str.rsplit(' ', 0).str.get(1)

top_ten_jobs
```

```python
#Sorting the rows in ascending order, based on number of companies with that particular job title and resetting the index using sort_values and sort_index
top_ten_jobs_sorted = top_ten_jobs.sort_values(by=['Company'], ascending=False)
top_ten_jobs_sorted = top_ten_jobs_sorted.sort_index(axis=1)
top_ten_jobs_sorted = top_ten_jobs_sorted[['JobTitle', 'Company']]

print("Number of job titles related to data science, across all companies")
top_ten_jobs_sorted.head(10)
```

#### How many job positions are there in each city?

#### Considering only the top 10 cities with the highest number of data analysis related jobs, let us consider the proportions of these job positions by city by plotting a pie chart:

#### What is the range of reviews posted for each job title and what does that mean?


```python
#Kernel density estimation - makes it easy to discern (central tendency, bimodality, skew)
plot4 = sns.displot(dsJobs_sorted, x='Reviews', kind='kde')
plt.title("Distribution of the Number of Reviews for Data Science Related Jobs") #including a title
plot4.fig.set_figheight(4) 
plot4.fig.set_figwidth(10) 
```

```python
#finding the most popular job titles - therefore, readily available roles
#counting the number of job titles
dsJobs_grouped_jobs = dsJobs_sorted.groupby(['JobTitle'], as_index=False, sort=False).sum()
dsJobs_grouped_jobs_1 = dsJobs_grouped_jobs.reset_index()
print("Frequency of job roles per title")
dsJobs_grouped_jobs_2 = dsJobs_grouped_jobs_1.drop(0)
top_ten_dsJobs_grouped_jobs_2 = dsJobs_grouped_jobs_2.head(10)
top_ten_dsJobs_grouped_jobs_2
```

This Kernel density estimation graph shows that individual job titles receive reviews ranging from 0 to 22,000. The jobs with high number of reviews are most likely entry-level positions, while the more specialized positions such as postdoctoral research have fewer reviews due to the high level of skill, education and experience required. Therefore, this means that fewer reviews should not deter candidates from reading through them in order to gain some insights about the role, because these reviews are just as valuable.
